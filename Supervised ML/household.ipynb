```python
import pandas as pd
```

```python
df=pd.read_csv('HouseholdIncome.csv')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
df['Education_Level']=l.fit_transform(df['Education_Level'])
```

```python
l1=LabelEncoder()
df['Occupation']=l1.fit_transform(df['Occupation'])
```

```python
l2=LabelEncoder()
df['Location']=l2.fit_transform(df['Location'])
```

```python
l3=LabelEncoder()
df['Marital_Status']=l3.fit_transform(df['Marital_Status'])
```

```python
l4=LabelEncoder()
df['Employment_Status']=l4.fit_transform(df['Employment_Status'])
```

```python
l5=LabelEncoder()
df['Homeownership_Status']=l5.fit_transform(df['Homeownership_Status'])
```

```python
l6=LabelEncoder()
df['Type_of_Housing']=l6.fit_transform(df['Type_of_Housing'])
```

```python
l7=LabelEncoder()
df['Gender']=l7.fit_transform(df['Gender'])
```

```python
l8=LabelEncoder()
df['Primary_Mode_of_Transportation']=l8.fit_transform(df['Primary_Mode_of_Transportation'])
```

```python
x=df.drop(['Income'],axis=1)
```

```python
y=df['Income']
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from xgboost import XGBRegressor
model=XGBRegressor()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
model.score(xtest,ytest)
```

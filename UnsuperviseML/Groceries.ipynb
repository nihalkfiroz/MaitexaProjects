```python
one_hot=pd.get_dummies(df['itemDescription']).astype('int')
df=df.join(one_hot)
```

```python
df
```

```python
products=df['itemDescription'].unique()
```

```python
df=df.drop('itemDescription',axis=1)
```

```python
df.head()
```

```python
records=df.groupby(['Member_number','Date'])[products[:]].apply(sum)
```

```python
def get_products_names(p):
    for product in products:
        if p[product]>0:
            p[product]=product
    return p
records=records.apply(get_products_names,axis=1)
```

```python
records
```

```python
records.values
```

```python
transaction=records.values
trans=[]
for i in transaction:
    trans.append(i[i!=0].tolist())
trans
```

```python
len(trans)
```

```python
rules=apriori(trans, min_support=0.0003,min_confidence=0.05,min_lift=3,min_length=2)
```

```python
association_result=list(rules)
len(association_result)
```

```python
association_result
```

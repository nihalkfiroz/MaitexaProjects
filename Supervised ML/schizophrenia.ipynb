```python
import pandas as pd
```

```python
df=pd.read_csv('schizophrenia_dataset.csv')
```

```python
df=df.rename(columns={'Hasta_ID':'PatientId','Yaş':'Age','Cinsiyet':'Gender','Eğitim_Seviyesi':'EducationLevel','Medeni_Durum':'MaritalStatus',
             'Meslek':'Occupation','Gelir_Düzeyi':'IncomeLevel','Yaşadığı_Yer':'LiveArea','Tanı':'Diagnosis','Hastalık_Süresi':'DiseseDuration',
             'Hastaneye_Yatış_Sayısı':'Hospitalizations','Ailede_Şizofreni_Öyküsü':'FamilyHistory','Madde_Kullanımı':'SubstanceUse','İntihar_Girişimi':'SuicideAttempt',
             'Pozitif_Semptom_Skoru':'PositiveSS','Negatif_Semptom_Skoru':'NegativeSS','GAF_Skoru':'GafScore','Sosyal_Destek':'SocialSupport',
             'Stres_Faktörleri':'StressFactors','İlaç_Uyumu':'MedicationAdherence'})
```

```python
df
```

```python
df.info()
```

```python
x=df.drop(['PatientId','Diagnosis'],axis=1)
```

```python
y=df['Diagnosis']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.tree import DecisionTreeClassifier
model1=DecisionTreeClassifier()
```

```python
model1.fit(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

```python
from sklearn.ensemble import RandomForestClassifier
model2=RandomForestClassifier(n_estimators=30,random_state=80)
model3=RandomForestClassifier(n_estimators=50,random_state=100)
```

```python
model2.fit(xtrain,ytrain)
model3.fit(xtrain,ytrain)
```

```python
import pandas as pd
df=pd.read_csv('energy_data.csv')
```

```python
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
```

```python
df['Building Type']=l.fit_transform(df['Building Type'])
```

```python
l1=LabelEncoder()
df['Day of Week']=l1.fit_transform(df['Day of Week'])
```

```python
x=df.drop(['Energy Consumption'],axis=1)
```

```python
y=df['Energy Consumption']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVR
model=SVR()
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
a=input('Enter Building Type')
a=l.transform([a])[0]
b=int(input('Enter Square Feet'))
c=int(input('Enter no.of occupants'))
d=int(input('Enter no.of Appliances used'))
e=int(input('Enter Avg.Temp'))
f=input('Enter Day of the week')
f=l1.transform([f])[0]
model.predict([[a,b,c,d,e,f]])
```

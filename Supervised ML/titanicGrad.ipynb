```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
df.info()
```

```python
df['Age'].fillna(30,inplace=True)
```

```python
df['Cabin'].backfill(inplace=True)
```

```python
df['Embarked'].backfill(inplace=True)
```

```python
df=df.drop(['Name'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
df['Sex']=l.fit_transform(df['Sex'])
```

```python
l1=LabelEncoder()
df['Ticket']=l1.fit_transform(df['Ticket'])
```

```python
l2=LabelEncoder()
df['Cabin']=l2.fit_transform(df['Cabin'])
```

```python
l3=LabelEncoder()
df['Embarked']=l3.fit_transform(df['Embarked'])
```

```python
x=df.drop(['Survived'],axis=1)
```

```python
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model=GradientBoostingClassifier()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(ypred,ytest)
cm
```

```python
print(classification_report(ypred,ytest))
```

```python
import pandas as pd
```

```python
df=pd.read_csv('loan_data.csv')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
df['person_gender']=l.fit_transform(df['person_gender'])
```

```python
l1=LabelEncoder()
df['person_education']=l1.fit_transform(df['person_education'])
```

```python
l2=LabelEncoder()
df['person_home_ownership']=l2.fit_transform(df['person_home_ownership'])
```

```python
l3=LabelEncoder()
df['loan_intent']=l3.fit_transform(df['loan_intent'])
```

```python
l4=LabelEncoder()
df['previous_loan_defaults_on_file']=l4.fit_transform(df['previous_loan_defaults_on_file'])
```

```python
x=df.drop(['loan_status'],axis=1)
```

```python
y=df['loan_status']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier()
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
model1=KNeighborsClassifier(n_neighbors=10)
model1.fit(xtrain,ytrain)
model1.score(xtest,ytest)
```

```python
model2=KNeighborsClassifier(n_neighbors=50)
model2.fit(xtrain,ytrain)
model2.score(xtest,ytest)
```

```python
model3=KNeighborsClassifier(n_neighbors=75)
model3.fit(xtrain,ytrain)
model3.score(xtest,ytest)
```
